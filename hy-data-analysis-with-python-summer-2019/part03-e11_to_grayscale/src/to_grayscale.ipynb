```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread("R:/Git/PythonDataAnalysisCourse/hy-data-analysis-with-python-summer-2019/part03-e11_to_grayscale/src/painting.png")
img0 = img.copy()
img1 = img.copy()
img2 = img.copy()
img.shape
```

```python
plt.imshow(img[:,:,])
```

```python
img0[:,:,[1,2]] *= 0
# img0[:,:,0] *= 1
plt.imshow(img0)
```

```python
img0[:,:,0] = 1
plt.imshow(img0)
```

1 is max, zero is min. So I maximized the entire first matrix and minimized the other two.
so the first (0) matrix encodes the red part of the image

```python
# img[:,:,0]
```

```python
img1[:,:,[0,2]] *= 0
plt.imshow(img1)
```

```python
img1[:,:,1] = 1
plt.imshow(img1)
```

so the second (1) matrix encodes the green part of the image

```python
img2[:,:,:2] *= 0
plt.imshow(img2)
```

```python
img2[:,:,2] = 1
plt.imshow(img2)
```
